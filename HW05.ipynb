## Q1. 寫一個函式，抓取 Google Finance 的股價歷史資料

```python
import requests
import pandas as pd
import io

def get_params(symbol,startdate,enddate):
    params = {'q': symbol,
              'startdate': startdate,
              'enddate': enddate,
              'output': "csv"
    }
    return params
    
def stockprice_google():
    symbol = input('股票代碼:')
    startdate = input('起始日期(ex.2016-01-01):')
    enddate = input('結束日期(ex.2016-01-01):')
    url = 'http://finance.google.com/finance/historical'
    
    r = requests.get(url, get_params(symbol,startdate,enddate))
    df = pd.read_csv(io.StringIO(r.text))
    return df
    
stockprice_google()
```
