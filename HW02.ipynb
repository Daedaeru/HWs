```python
#載入模組
import os

#定義套件(要記得加冒號
def main():
    #print出想要顯示的文字
    print ('Hello word!')               #()的差別
    print("This is Allice's greeting.") #體會雙引號
    print('This is Bob\'s greeting.')   #體會單引號(\斜線會不見)
    
    #用foo套件放入參數5跟10
    foo(5,10)
    
    #print出10個=，注意括號的位置
    print (('=')*10)
    #除了文字也可以加指令print在一起，這裡的os.getcwd是os的
    print (('Current working directory is ')+ os.getcwd())
    
    counter = 0       #指定變數，例如A=1
    counter += 1      #counter=counter+1(好神奇的語法
    
    #設迴圈，需要加, end=''
    food=['apples','orange','cats XDDDDD']
    for i in food :
        print ((' I like to eat ' )+ i , end='')
    
    #可以發現其實是從0開始數(用range列出，數10個所以到數字9
    print('Count to ten:')
    for x in range(10) :
        print (x)
    
#定義套件並設置參數
def foo (paraml, secondParam) :
    res = paraml + secondParam
    
    #這裡很好玩，可以隨結果改變print出的字
    print('%s plus %s is egual to %s' % (paraml, secondParam,res))
    #下面兩種寫法是只能固定的(不像上面可以浮動
   # print('{a} plus {b} is egual to {c}'.format(a='paraml',b='secondParam',c='res') ) 
   # print('%(a)s plus %(b)s is egual to %(c)s' % 
   #       ({'a' : 'paraml', 'b' : 'secondParam', 'c' : 'res'}))

    #條件判斷(
    if res < 50 :
        print('foo')
        
    elif (res >= 50 ) and ((paraml == 42)or(secondParam == 24)):
        print('bar')
    
    else:
        print('moo')
        
    return res                    #最後顯示res
    #This is a one-line comment.  #只能打一行
    '''A multi-line string,                 
    but can also be a multi-line comment'''  
    '''可以
    很多行註解，很棒'''
    
#天啊我終於搞懂name跟main了!(希望真的是
if __name__ == '__main__':
    main()

    '''假設有兩個.py,我們分別叫它們A.py跟B.py
    假設B.py自己執行，則__name__就會是main；
    若今天B.py被當作模組import到A.py，這個時候B.py的__name__就是B。
    
    所以__name__是變數，__main__或名字是值，
    這兩個就是用來判別是模組或是直接執行，其實不會改名字嘛!厚'''
```
